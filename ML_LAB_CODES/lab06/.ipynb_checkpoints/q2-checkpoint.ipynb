```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv')
x = df['experience']
y = df['salary']
```

```python
alpha = 0.01
b0 = 0.0
b1 = 0.0
error = []
beta = []
for i in range(5):
    for j in range(12):
        y_pred = ((b1*x[j])+b0)
        delta = y_pred - y[j]
        error.append(delta)
        b0 = b0 - (alpha*delta)
        b1 = b1 - (alpha*delta*x[j])
        beta.append(b1)
```

```python
print(b0,b1)
```

```python
import numpy as np
mse = (np.sum(y-(b0 + (b1*x))))**2/12
mse
```

```python
plt.scatter(x,y)
plt.plot(x,(b0 + (b1*x)),color = 'red')
```

```python
#from sklearn.model_selection import train_test_split
x = np.array(x)
x = x.reshape(-1,1)
from sklearn.linear_model import SGDRegressor
reg = SGDRegressor(max_iter=60, alpha=alpha)
model = reg.fit(x, y)
y_pred = model.predict(x)
```

```python
print(model.intercept_)
print(model.coef_)
y = np.array(y)
```

```python
mse = np.sum(y - ((model.intercept_) + (model.coef_*x)))**2/12
mse
```

```python
plt.scatter(beta,error)
plt.plot(beta,error,color = 'red')
```
